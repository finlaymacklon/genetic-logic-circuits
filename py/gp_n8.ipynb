```python
import pandas as pd
import numpy as np

from gp_boolean.gp_search import genetic_programming_search
from gp_boolean.gp_node import X
```

```python
"""
Set up the problem to be solved
"""

# set headers for truth table (inputs and output)
#table_headers = [
#    ['in', 'in', 'in', 'in', 'in', 'in', 'in', 'in', 'in','in', 'in',    'out'],
#    ['S0', 'S1', 'S2',  'A',  'B',  'C',  'D',  'E',  'F', 'G',  'H',     '']
#]
table_headers = [
    ['in', 'in', 'in', 'in', 'in', 'in', 'in', 'in', 'in','in', 'in',    'out'],
    ['S0', 'S1', 'S2',  'A',  'B',  'C',  'D',  'E',  'F', 'G',  'H',     '']
]
# set truth table data
#table_data = [
#    [   0,    0,    0,    1,    X,    X,   X,    X,    X,   X,    X,      1],
#    [   0,    0,    1,    X,    1,    X,   X,    X,    X,   X,    X,      1],
#    [   0,    1,    0,    X,    X,    1,   X,    X,    X,   X,    X,      1],
#    [   0,    1,    1,    X,    X,    X,   1,    X,    X,   X,    X,      1],
#    [   1,    0,    0,    X,    X,    X,   X,    1,    X,   X,    X,      1],
#    [   1,    0,    1,    X,    X,    X,   X,    X,    1,   X,    X,      1],
#    [   1,    1,    0,    X,    X,    X,   X,    X,    X,   1,    X,      1],
#    [   1,    1,    1,    X,    X,    X,   X,    X,    X,   X,    1,      1],
#]
table_data = [
    [   0,    0,    0     1,    X,    X,   X,    X,    X,   X,    X,      1],
    [   0,    0,    1,    X,    1,    X,   X,    X,    X,   X,    X,      1],
    [   0,    1,    0,    X,    X,    1,   X,    X,    X,   X,    X,      1],
    [   0,    1,    1,    X,    X,    X,   1,    X,    X,   X,    X,      1],
    [   1,    0,    0,    X,    X,    X,   X,    1,    X,   X,    X,      1],
    [   1,    0,    1,    X,    X,    X,   X,    X,    1,   X,    X,      1],
    [   1,    1,    0,    X,    X,    X,   X,    X,    X,   1,    X,      1],
    [   1,    1,    1,    X,    X,    X,   X,    X,    X,   X,    1,      1],
]
# create dataframe for truth table
truth_table = pd.DataFrame(
    columns = table_headers,
    data = table_data
)
# show truth table
truth_table.head(10)
```

```python
"""
Perform the search
"""

kwargs = {
    'max_depth': 6, 
    'size_pop': 100, 
    'size_mate': 10,
    'num_gens': 100, 
    'p_mutate': 0.5, 
    'p_recombine': 0.5,
    'w_correct': 1,
    'w_size': 0.1
}

all_gp = []
all_times = []
all_pcts = []
all_sizes = []

for i in range(10):
    gp, walltime, pct, size = genetic_programming_search(truth_table, **kwargs)
    all_gp.append(gp)
    all_times.append(walltime)
    all_pcts.append(pct)
    all_sizes.append(size)
    gp.print_tree()
    print('{:.0f}% correct'.format(100*pct))
    print('Walltime = {:.3f} seconds'.format(walltime))
    print('Size = {} gates'.format(size))
    print('------------------------------')
```
